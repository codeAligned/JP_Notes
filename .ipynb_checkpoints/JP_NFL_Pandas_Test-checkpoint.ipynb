```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import math
%matplotlib inline
```

```python
df14 = pd.read_csv('/Users/jpw/Desktop/jp_nfl_pandas_tester.csv')
```

```python
#Want to make row index(0) the Header row
df14.columns = df14.iloc[0]
df14.head()
```

```python
#Now need to drop row index(0)
df14.drop(df14.index[[0]], inplace=True)
df14.reindex(df14) #Can use this to reindex the DF (to maintain 0-indexed structure)
df14.head()
```

```python
#clean up col names by replacing spaces with underscores -- needs to handle floats
cols = df14.columns.tolist()
cols = [str(col) for col in cols] #convert all header names to str first
cols = [col.replace(' ', '_') for col in cols]
cols = [''.join([i if 32 < ord(i) < 126 else "" for i in col]) for col in cols] #remove non-ASCII chars in headers

df14.columns = cols
df14.head()
```

```python
#drop specific columns by index - example
df14.drop(df14.columns[[100, 103, 106]], axis=1, inplace=True)
df14.head()
```

```python
#drop a range of unwanted columns by index - example; this is usually not the best way to approach this.  Better to 
#simply make a new df using a mask of the desired cols from the original df
df14.drop(df14.columns[89:], axis=1, inplace=True)
df14.head()
```

```python
#View all column names
print len(df14.columns)
```

```python
#Create a new DF just for plotting Weekly Port%, by index since multiple columns are named 'week x'
#len(df14.columns)
dfplot = df14.ix[0:32, [67:89]
dfplot
```

```python
#Try to 'reindex' rows as the names of the teams - failing
dfplot.rows = dfplot.ix[:,'TEAM']
dfplot
```

```python
#convert data to floats instead of str

dfplot = dfplot.apply(lambda x: pd.to_numeric(x, errors='ignore'))
dfplot.head()
```

```python
#SB teams weekly data
sea_port = np.array(dfplot.ix[28, 1:])
sea_port[-4] = sea_port[-5]
ne_port = np.array(dfplot.ix[21, 1:])
ne_port[-4] = ne_port[-5]
ne_port[-1] = .737
```

```python
#Use Plotting DF for plotting in MPL
weeks = np.arange(1, 22, 1)

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

ax.plot(weeks, sea_port, color='#66c010', marker='o', lw=2, label='Seattle')
ax.plot(weeks, ne_port, color='#011582', marker='o', lw=2, label='New England')
ax.set_title('2014 Super Bowl Teams')
ax.set_ylabel('Pythagenport Rating')
ax.set_xlabel('Week')
ax.grid(True)
plt.legend(loc='best')
plt.axis([1, 22, 0, 1])



# Adjust the tick positions and labels
#grid.set(xticks=np.arange(5), yticks=[-3, 3],
 #        xlim=(-.5, 4.5), ylim=(-3.5, 3.5))
#plt.xscale('log')
plt.show()
```
